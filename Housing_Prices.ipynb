### Encoding

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler, MinMaxScaler
from sklearn.ensemble import RandomForestRegressor

label_encoder = LabelEncoder()

categorical_columns = ['mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'prefarea', 'furnishingstatus']

for column in categorical_columns:
    df[column] = label_encoder.fit_transform(df[column])
```

```python
df.head()
```

### Data Splitting

```python
x = df.drop('price', axis=1)
y = df['price']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
scaler = MinMaxScaler()

x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
x_train
```

### Modelling

```python
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.svm import SVR
from sklearn.linear_model import Lasso, Ridge
from lightgbm import LGBMRegressor
import xgboost as xgb

rfr = RandomForestRegressor().fit(x_train, y_train)
gbr = GradientBoostingRegressor().fit(x_train, y_train)
svr = SVR().fit(x_train, y_train)
```

```python
from sklearn.metrics import root_mean_squared_error, mean_squared_error, r2_score

def evaluate_model(model, x_test, y_test):
    y_pred = model.predict(x_test) # Use x_test for prediction
    results = {
        'RMSE': root_mean_squared_error(y_test, y_pred),
        'R2 Score': r2_score(y_test, y_pred,)
    }
    return results

# Mengevaluasi setiap model
results = {'Random Forest Regressor': evaluate_model(rfr, x_test, y_test),
           'Gradient Boosting': evaluate_model(gbr, x_test, y_test),
           'Support Vector Machine': evaluate_model(svr, x_test, y_test)}

summary_df = pd.DataFrame(columns=['RMSE', 'R2 Score'])

rows = []
for model_name, metrics in results.items():
    rows.append({
        'Model': model_name,
        'RMSE': metrics['RMSE'],
        'R2 Score': metrics['R2 Score']
    })

summary_df = pd.DataFrame(rows)
print(summary_df)
```

### Model Inference

```python
# Select a random sample from the original dataframe
random_sample = df.sample(1)

# Separate features (x) and target (y) for the random sample
random_sample_x = random_sample.drop('price', axis=1)
random_sample_y = random_sample['price']

# Scale the features of the random sample using the same scaler fitted on the training data
random_sample_x_scaled = scaler.transform(random_sample_x)

print("Random Sample (Original Data):")
display(random_sample)
```

```python
# Make predictions using each trained model
rfr_pred = rfr.predict(random_sample_x_scaled)
gbr_pred = gbr.predict(random_sample_x_scaled)
svr_pred = svr.predict(random_sample_x_scaled)

# Create a DataFrame to display the results
inference_results = pd.DataFrame({
    'Model': ['Random Forest Regressor', 'Gradient Boosting', 'Support Vector Machine'],
    'Predicted Price': [rfr_pred[0], gbr_pred[0], svr_pred[0]],
    'Actual Price': random_sample_y.iloc[0]
})

display(inference_results)
```

```python
# Visualize predictions vs actual values for each model
plt.figure(figsize=(15, 5))

# Random Forest Regressor
plt.subplot(1, 3, 1)
plt.scatter(y_test, rfr.predict(x_test), alpha=0.5)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=2)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Random Forest Regressor: Actual vs Predicted Price")

# Gradient Boosting Regressor
plt.subplot(1, 3, 2)
plt.scatter(y_test, gbr.predict(x_test), alpha=0.5)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=2)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Gradient Boosting: Actual vs Predicted Price")

# Support Vector Machine
plt.subplot(1, 3, 3)
plt.scatter(y_test, svr.predict(x_test), alpha=0.5)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=2)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Support Vector Machine: Actual vs Predicted Price")

plt.tight_layout()
plt.show()
```
